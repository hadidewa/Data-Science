```python
def cek_bmi(bmi):
    result = 'Low BMI'
    if bmi >= 26:
        result = 'High BMI'
    return result
```

```python
for i in range(len(data)):
    bmi = data.loc[i, 'bmi']
    data.loc[i, 'cat_bmi_loop'] = cek_bmi(bmi)
```
